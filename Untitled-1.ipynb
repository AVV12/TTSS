```python
import pandas as pd
from sktime.utils.plotting import plot_series
import seaborn as sns
from sktime.transformations.series.boxcox import LogTransformer
from sktime.forecasting.base import ForecastingHorizon
from sktime.forecasting.model_selection import temporal_train_test_split
from sktime.utils.plotting import plot_correlations
from sktime.forecasting.arima import AutoARIMA
import numpy as np
from sktime.performance_metrics.forecasting import mean_absolute_percentage_error
from sktime.performance_metrics.forecasting import MeanSquaredError
```

```python
df_app = pd.read_csv('aapl_ok.csv', delimiter=';')
df_app['fecha'] = pd.to_datetime(df_app['fecha'])
df_app = df_app.set_index('fecha')
df_app['Ingresos']=df_app['Ingresos'].str.replace(',', '.')
df_app['Ingresos']=pd.to_numeric(df_app['Ingresos'], errors= 'coerce')
df_app = df_app.resample('q').last()
df_app.head()
```

```python
df_app_m = df_app.drop('Trimeste', axis=1)
```

```python
df_app_q=df_app_m['Ingresos'].astype('float64').to_period('Q')
df_app_q.tail()
```

```python
# Aplicamos el Log Transformer
transformer = LogTransformer()
log_app_ts= transformer.fit_transform(df_app_q)
log_app_ts.tail()
```

```python
plot_correlations(log_app_ts)
```

```python
y_train, y_test = temporal_train_test_split(y =df_app_q, test_size=8)
log_y_train, log_y_test = temporal_train_test_split(y =log_app_ts, test_size=8)
plot_series(y_train, y_test, labels=["Train", "Test"])
# Time Series Size
print(y_train.shape[0], y_test.shape[0])
# Consigo printear el Train pero el test no puedo sacarlo
```

```python
fh = np.arange(len(y_test)) + 1  # forecasting horizon
fh
```

```python
forecaster = AutoARIMA(sp=4,suppress_warnings=True)
forecaster.fit(log_y_train)
```

```python
print(forecaster.summary())
```

```python
log_y_pred = forecaster.predict(fh)
log_y_pred
```

```python
#  Auto ARIMA Forecast
np.exp(log_y_pred)
```

```python
# Forecast Accuracy MAPE & MAPE con el original
mean_absolute_percentage_error(log_y_test, log_y_pred), mean_absolute_percentage_error(y_test, np.exp(log_y_pred))
```

Viendo los dos errores vemos que son muy parecidos y sigue siendo un error pequeño con el que podemos trabajar.

Con el fin de observar con mayor seguridad la posible evolución ploteamos

```python
plot_series(y_train, np.exp(log_y_pred),y_test, labels=["App", "App pred", "App REAL"])
```

```python
# Forecast Accuracy Plot
# Zoom 2013 -2021
plot_series(y_train["2013":], np.exp(log_y_pred),y_test, labels=["App", "App pred", "App REAL"])
```
